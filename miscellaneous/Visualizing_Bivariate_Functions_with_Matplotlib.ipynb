In this notebook, we present three simple bivariate functions $f_1$, $f_2$, and $f_3$ and three points $P_1$, $P_2$, $P_3$ such that


*   $P_1$ is a minimum of $f_1$.
*   $P_2$ is a maximum of $f_2$.
*   $P_3$ is neither a minimum nor maximum of $f_3$, but the gradient of $f_3$ at $P_3$ is the zero vector.

Then, we use *Matplotlib* to visualize them.

## 1. Import Statements

---



```
# Import Statements

from mpl_toolkits import mplot3d

%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Bivariate Functions

---



First, let us define the three bivariate functions.

*   $f_1 = x^2y^2$
*   $f_2 = -x^2\sqrt{|y|}$
*   $f_3 = xy - x$

Further, let us define the points.

*   The point $P_1 = (-1, 0)$ is a minimum of $f_1$.
*   The point $P_2 = (-1, 0)$ is a maximum of $f_2$.
*   The point $P_3 = (1, 0)$ is neither a minimum nor maximum of $f_3$, but the gradient of $f_3$ at $P_3$ is the zero vector.

### 2.1. Function Definitions

```
def f1(x, y):
  return (x ** 2) * (y ** 2)

def f2(x, y):
  return -(x ** 2) * np.sqrt(np.abs(y))

def f3(x, y):
  return (x * y) - x
```

### 2.2. Data Generation and Initialization

```
# Generate random values for x and y.
x = np.linspace(-25, 25, 50)
y = np.linspace(-25, 25, 50)

# Create a grid using x and y.
X, Y = np.meshgrid(x, y)

# Call the three function using the grid.
Z1 = f1(X, Y)
Z2 = f2(X, Y)
Z3 = f3(X, Y)

# Create the points.
P1 = [-1, 0, f1(-1, 0)]
P2 = [-1, 0, f2(-1, 0)]
P3 = [1, 0, f3(1, 0)]
```

## 3. Visualization Functions

---



```
def plot_contour3D(f, P):

  '''
  Function for plotting a 3D Contour Model of a function f and a given point P.
  '''

  # Initialize the figure and the axes. Set to 3-D Contour Plot.
  fig = plt.figure()
  ax = plt.axes(projection='3d')
  ax.contour3D(X, Y, f, 1000, cmap='RdGy')

  # Set the axis labels.
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('z')
  ax.view_init(60, 45)

  # Plot the point.
  Px = P[0]
  Py = P[1]
  Pz = P[2]
  ax.plot([Px], [Py], [Pz], 'go', markersize = 10)

  plt.tight_layout()

def plot_contour2D(f, P):

  '''
  Function for plotting a 2D Contour Model of a function f.
  '''

  plt.contour(X, Y, f, 50, cmap='RdGy')

  # Plot the point.
  Px = P[0]
  Py = P[1]
  plt.plot(Px, Py, 'go', markersize = 10)

  plt.tight_layout()

def plot_contourf(f, P):

  '''
  Function for plotting a filled 2D Contour Model of a function f.
  '''

  plt.contourf(X, Y, f, 100, cmap='RdGy')

  # Plot the point.
  Px = P[0]
  Py = P[1]
  plt.plot(Px, Py, 'go', markersize = 10)

  plt.tight_layout()
```

## 4. Plots

---



### 4.1. Plots for $f_1 = x^2y^2$

```
plot_contour3D(Z1, P1)
```

```
plot_contour2D(Z1, P1)
```

```
plot_contourf(Z1, P1)
```

### 4.2. Plots for $f_2 = -x^2\sqrt{|y|}$

```
plot_contour3D(Z2, P2)
```

```
plot_contour2D(Z2, P2)
```

```
plot_contourf(Z2, P2)
```

### 4.3. Plots for $f_3 = xy - x$

```
plot_contour3D(Z3, P3)
```

```
plot_contour2D(Z3, P3)
```

```
plot_contourf(Z3, P3)
```
